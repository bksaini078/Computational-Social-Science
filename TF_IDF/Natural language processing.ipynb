 Convert the input text to a list of sentences. Then, compute the number of sentences in the given Text.

```
import os 
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
import nltk
import spacy
ps = PorterStemmer()

nlp = spacy.load("en_core_web_sm")
nltk.download('stopwords')
#removing stop words for 
stopWords = set(stopwords.words("english"))

path ='/content/drive/My Drive/CSS/Article.txt'
file = open(path, "r") 
content= file.readlines()
content
```

Calculate the frequency of words in each sentence: the output should be a dictionary where each key is a sentence and the value is also a dictionary of word frequency.

```
doc= nlp(content[0])
sentences = (list(doc.sents))
str(sentences[0]).lower()
# maving sentences in lower case 
sentences= [str(sent).lower() for sent in sentences ]
total_sentence=len(sentences )
total_sentence
```

```
total_sentence
w_sent_freq_dict={}
bad_chars = [ '-','|',':', '!', "*",'?','[',']','(',')','\xa0',',','.','+','"','','\\'] 
for sent in sentences:
    word_list=[]
    # sent= nlp(sent)
    sent = ''.join(i for i in sent if not i in bad_chars) 
    sent= nlp(sent)
    sent= " ".join([token.lemma_ for token in sent])
    word_list= sent.split(' ')
    w_freq_dict={}
    
    for w in word_list: 
        # w= ps.stem(w)
        if w in stopWords:
            continue
        w_freq_dict[w]= word_list.count(w)
    w_sent_freq_dict[sent]= w_freq_dict

w_sent_freq_dict
```

3. Calculate Term frequency for each word in a sentence:
**TF(word)=(Number of times term “word” appears in a sentence) / (Total number of terms in the sentence)**

```
tf_matrix={}
for key in w_sent_freq_dict:
    w_freq_k= w_sent_freq_dict.get(key)
    # print(w_freq_k)
    for k in w_freq_k:
        # print(len(w_freq_k))
        w_freq_k[k]= (w_freq_k[k]/len(w_freq_k))
        # print(len(w_freq_k),w_freq_k.get(k))
    tf_matrix[key]=w_freq_k
    # for k in w_freq:
#w_sent_freq_dict
tf_matrix
```

For each word compute how many sentences contain that word.
Calculate IDF for each word in a sentence.

IDF(word) = log e(Total number of sentences/ number of sentences with term word in it) 

```
#creating document frequency dictionary
word_df={}

for sent, word_tf in tf_matrix.items():
    for word, tf in word_tf.items():
        if word  in word_df:
            word_df[word]+=1
        else:
            word_df[word]=1
```

```
# creating inverse document frequency matrix 
import math
word_idf_matrix={}
for sent,word_tf in tf_matrix.items():
    word_idf ={}
    for word,tf in word_tf.items():
        word_idf[word]= math.log(total_sentence/word_df[word])
    word_idf_matrix[sent]= word_idf
word_idf_matrix
```

Compute the TF-IDF for each word in each sentence.
Use the TF-IDF computed in (7) and give a weight for each sentence.

Threshold: compute the average sentence weight



```
# tf idf calculation
tf_idf_matrix={}
for (sent1,wordl1),(sent2,wordl2) in zip(tf_matrix.items(),word_idf_matrix.items()):
    word_tfidf={}
    for (word1,tf),(word2,idf) in zip(wordl1.items(),wordl2.items()):
        word_tfidf[word1]= tf*idf
    tf_idf_matrix[sent1]=word_tfidf
tf_idf_matrix        
```

```
# giving weight for each sentence 
sentence_weight={}
# this one to calculate threshold of the value 
thres_cal=[]
for sent, word_tfidf in tf_idf_matrix.items():
    # just adding the tfidf
    tf_idf_score=0
    for word, tf_idf in word_tfidf.items():
        tf_idf_score += tf_idf
    sentence_weight[sent]= tf_idf_score
    thres_cal.append(tf_idf_score)

threshold= sum(thres_cal)/len(thres_cal)

sentence_weight,threshold
```

Generate the summary : select a sentence for summarization if the weight of the sentence exceeds the threshold.

```
summary=''
for sents,weight in sentence_weight.items():
    if weight > threshold:
        print(sents)
        summary += " " + sents 
        
 
# summary
```
